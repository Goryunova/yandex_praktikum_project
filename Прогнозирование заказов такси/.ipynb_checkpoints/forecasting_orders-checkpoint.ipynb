#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Вам нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

## Подготовка

```python
# загружаем необходимые библиотеки
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import time

from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.metrics import make_scorer
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import OrdinalEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import TimeSeriesSplit
from sklearn.model_selection import train_test_split
from statsmodels.tsa.seasonal import seasonal_decompose

import warnings
warnings.filterwarnings('ignore')
```

## Обучение

Создадим дополнительные признаки, которые могут помочь в обучении модели. Т.к. стоит задача краткосрочного прогноза (на 1 час вперед), то такие признаки как день недели, месяц и год необходимо не включать. Добавим лаги на 24 часа, скользящее среднее с горизонтом в 7 часов, час и день.

```python
def make_features(data, max_lag, rolling_mean_size):
    data_new = data.copy()
    data_new['hour'] = data.index.hour
    data_new['dayofweek'] = data.index.dayofweek
    data_new['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
    for lag in range(1, max_lag + 1):
        data_new['lag_{}'.format(lag)] = data_new['num_orders'].shift(lag)
    return data_new
```

```python
data_1 = make_features(data, 2, 24)
data_1.head(25)
```

```python
# Разделим данные на обучающую и тестовую выборки
train, test = train_test_split(data_1, shuffle=False, test_size=0.1)
# Проверим корректность деления
print(train.index.min(), train.index.max())
print(test.index.min(), test.index.max())
```

```python
# Деление корректное. Удалим пропуски.
train.dropna(inplace=True)
train.info()
```

```python
# Создадим обчающий и целевой признаки
target_train = train['num_orders']
features_train = train.drop('num_orders', axis=1)
target_test = test['num_orders']
features_test = test.drop('num_orders', axis=1)
```

```python
# Создадим функцию для вычисления RMSE.
def rmse_score(target, predictions):
    rmse = mean_squared_error(target, predictions) ** 0.5
    return rmse
scorer = make_scorer(rmse_score, greater_is_better=False)
```

Обучим четыре модели: линейная регрессия, случайный лес, LightGBM, CatBoost. Поиск гиперпараметров будем проводить с помощью методов RandomizedSearchCv и TimeSeriesSplit.

### Линейная регрессия

```python
tscv = TimeSeriesSplit(n_splits=8)
```

```python
model = LinearRegression()
parameters = {'fit_intercept': [True, False], 'normalize': [True, False]}
grid = RandomizedSearchCV(model, parameters, cv=tscv, scoring=scorer, verbose=5, n_jobs=-1)
grid.fit(features_train, target_train)
print('RMSE линейной регрессии=', -grid.best_score_)
print('Наилучшие подобранные параметры для линейной регрессии', grid.best_params_)
```

Наилучший RMSE=31.36 получен для параметров: fit_intercept=False, normalize=True.

### Случайный лес

```python
forest = RandomForestRegressor(random_state=12345, n_estimators=80)
forest_params = {'max_features': ['auto', 'sqrt', 'log2'], 'max_depth': range(1, 15)}
forest_grid = RandomizedSearchCV(forest, forest_params, cv=tscv, scoring=scorer, verbose=5, n_jobs=-1)
forest_grid.fit(features_train, target_train)
print('RMSE для случайного леса составляет', -forest_grid.best_score_)
print('Наилучшие подобранные параметры для случайного леса', forest_grid.best_params_)
```

Наилучший RMSE=24.63 получен для параметров: max_features=auto, max_depth=13.

### Light GBM

```python
gbm = LGBMRegressor(random_state=12345)
gbm_params = {'n_estimators': [100, 500], 'num_leaves': [31, 41], 'learning_rate': [0.04, 0.06]}
gbm_grid = RandomizedSearchCV(gbm, gbm_params, cv=tscv, verbose=5, scoring=scorer)
gbm_grid.fit(features_train, target_train)
print('RMSE для LGBM составляет', -gbm_grid.best_score_)
print('Наилучшие подобранные параметры для LGBM', gbm_grid.best_params_)
```

Наилучший RMSE=24.43 получен для параметров: num_leaves=41, n_estimators=100, learning_rate=0.04.

## CatBoost

```python
cbr =CatBoostRegressor(loss_function='RMSE')
cbr_params = {'learning_rate': [0.03, 0.1, 0.5], 'iterations': [150, 500]}
cbr_grid = RandomizedSearchCV(cbr, cbr_params, cv=tscv, verbose=5, scoring=scorer)
cbr_grid.fit(features_train, target_train)
print('RMSE для CatBoostRegressor составляет', -cbr_grid.best_score_)
print('Наилучшие подобранные параметры для CatBoostRegressor', cbr_grid.best_params_)
```

Наилучший RMSE=24.60 получен для параметров: learning_rate=0.1, iterations=150.

**Вывод:** Обучены и подобраны гиперпараметры для 4 различных моделей: линейная регрессия, случайный лес, LightGBM, CatBoost. Для всех моделей была применена кросс-валидация с применением метода TimeSeriesSplit.
Целеовое значение метирки RMSE (<48) достигнуто всеми четырьмя моделями. наихудший результат быть получен с помощью линейной регрессии (31.36), наилучший - с помощью LGBM (24.43).
Проверим модели с подобранными гиперпараметрами на тестовой выборке.

## Тестирование

```python
# Создадим таблицу для хранения результатов
d = {'model': [], 'RMSE': [], 'fit_time': [], 'predict_time': []}
results = pd.DataFrame(data=d)
results
```

```python
# Создадим функцию для добавления результатов в таблицу
def model_res(model_name, model, features_train, target_train, features_test, target_test):
    rmse, fit_time, pred_time = ml(model, features_train, target_train, features_test, target_test)
    results.loc[len(results)] = [model_name, rmse, fit_time, pred_time]
    if len(results) > 0:
        results.drop_duplicates(inplace=True)
```

```python
# Создадим функцию для расчета необходимых параметров
def ml(model, features_train, target_train, features_test, target_test):
    start_fit_time = time.time()
    model.fit(features_train, target_train)
    stop_fit_time = time.time()
    fit_time = stop_fit_time - start_fit_time
    
    start_pred_time = time.time()
    predictions = model.predict(features_test)
    stop_pred_time = time.time()
    pred_time = stop_pred_time - start_pred_time
    
    rmse = mean_squared_error(target_test, predictions) ** 0.5
    return rmse, fit_time, pred_time
```

```python
# Создадим функцию для вывода графиков
def plot_test(model, model_name):
    plt.figure(figsize=(15, 5))
    plt.title('Сравнение реальных значений и предсказанных с помощью модели' + model_name)
    plt.plot(target_test, label='Test values')
    predictions = model.predict(features_test)
    predictions = pd.Series(predictions, index=target_test.index)
    plt.plot(predictions, label='Prediction values')
    plt.legend(loc='upper left')
    plt.grid(True)
```

```python
# Линейная регрессия
lr = grid.best_estimator_
model_res('LinearRegression', lr, features_train, target_train, features_test, target_test)
print(results)
plot_test(lr, 'LinearRegression')
```

На тестовой выборке линейная регрессия показала намного меньший результат и вывалилась за предельное значение метирки. график показывает, что, в основном, линейная регрессия больше всего ошибается в экстремумах.

```python
# Случайный лес
rf = forest_grid.best_estimator_
model_res('RandomForestRegressor', rf, features_train, target_train, features_test, target_test)
print(results)
plot_test(rf, 'RandomForestRegressor')
```

Метрика случайного леса получилась лучше, чем у линейной регрессии и вошла в допустимый интервал.

```python
# Light GBM
lgbm = gbm_grid.best_estimator_
model_res('LGBMRegresor', lgbm, features_train, target_train, features_test, target_test)
print(results)
plot_test(lgbm, 'LGBMRegressor')
```

```python
Метрика RMSE у LightGBM чуть лучше чем у случайного леса, а скорость предсказания в 3 раза выше.
```

```python
# CatBoost
cb = cbr_grid.best_estimator_
model_res('CatBoostRegressor', cb, features_train, target_train, features_test, target_test)
print(results)
plot_test(cb, 'CatBoostRegression')
```

**Вывод:** По итогу тестирования все модели, кроме линейной регрессии, подошли под требования по поставленной метирке (RMSE < 48), тем не менее каждая из моделей оказалась переобученной (или недообученной), т.к. результаты по тестовой выборке ухудшились почти в 1.5 раза для каждой модели по сравнению с обучающей.
С лучшими по итогу параметрами (метрика + скороть обучения) оказался LGBMRegressor, данную модель и подобранные ею гиперпараметры принимаем за основную и передаем заказчику для предсказания заказов такси в следующий час.
